```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns  
import math

df = pd.read_csv('/Volumes/Extreme Pro/DSAN-5200/final-project/T_DB1B_MARKET_CLEAN.csv')   
```

```python
df.head()
```

```python
df = df[['ORIGIN', 'DEST', 'MARKET_FARE', 'MARKET_DISTANCE']]
df.head()
```

```python
df.to_csv('/Volumes/Extreme Pro/DSAN-5200/final-project/origin_dest_price.csv', index=False)
```

```python
df["ROUTE"] = df["ORIGIN"] + "-" + df["DEST"]
```

```python
route_summary = (
    df.groupby("ROUTE")
    .agg(
        avg_fare=("MARKET_FARE", "mean"),
        median_fare=("MARKET_FARE", "median"),
        min_fare=("MARKET_FARE", "min"),
        max_fare=("MARKET_FARE", "max"),
        std_fare=("MARKET_FARE", "std"),
        avg_distance=("MARKET_DISTANCE", "mean"),
        num_observations=("MARKET_FARE", "count")
    )
    .reset_index()
)
```

```python
route_summary[["ORIGIN", "DEST"]] = route_summary["ROUTE"].str.split("-", expand=True)
```

```python
route_summary.to_csv('/Volumes/Extreme Pro/DSAN-5200/final-project/route_summary.csv', index=False)
```
